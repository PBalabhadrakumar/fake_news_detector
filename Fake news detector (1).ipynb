```python
data_merge=pd.concat([data_fake,data_true],axis=0)
data_merge.head(10)
```

```python
data=data_merge.drop(['title','subject','date'],axis=1)
```

```python
data.isnull().sum()
```

```python
data=data.sample(frac=1)
```

```python
data.head()
```

```python
data.reset_index(inplace=True) 
data.drop(['index'], axis=1, inplace=True)  
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
    text = text.lower()  
    text = re.sub(r'\[.*?\]', '', text) 
    text = re.sub(r'\\W', ' ', text) 
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  
    text = re.sub(r'<.*?>+', '', text)  
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)  
    text = re.sub(r'\n', '', text) 
    text = re.sub(r'\w*\d\w*', '', text)  
    text = re.sub(r'\s+', ' ', text).strip() 
    return text
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
x=data['text']
y=data['class']
```

```python
x_train ,x_test, y_train, y_test= train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


vectorizer = TfidfVectorizer()


xv_train = vectorizer.fit_transform(x_train)


xv_test = vectorizer.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression 

LR=LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT=  DecisionTreeClassifier()
DT.fit(xv_train,y_train)
```

```python
pred_dt=DT.predict(xv_test)
```

```python
DT.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GB=GradientBoostingClassifier(random_state=0)
GB.fit(xv_train,y_train)
```

```python
pred_gb=GB.predict(xv_test)
```

```python
GB.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF=RandomForestClassifier(random_state=0)
RF.fit(xv_train,y_train)
```

```python
pred_rf=RF.predict(xv_test)
```

```python
RF.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_rf))
```

```python
def output_label(n):
    if n==0:
        return "Fake News"
    elif n==1:
        return "Not Fake News"

def manual_testing(news):
    testing_news={"text":[news]}
    news_def_test=pd.DataFrame(testing_news)
    news_def_test["text"]=news_def_test["text"].apply(wordopt)
    news_x_test=news_def_test["text"]
    news_xv_test=vectorizer.transform(news_x_test)
    pred_LR=LR.predict(news_xv_test)
    pred_DT=DT.predict(news_xv_test)
    pred_GB=GB.predict(news_xv_test)
    pred_RF=RF.predict(news_xv_test)

    return print("\n\nLR Prediction:{}\nDT Prediction:{}\nGB Prediction:{}\nRF Prediction:{}".format(output_label(pred_LR[0]),
                                                                                                     output_label(pred_DT[0]),
                                                                                                     output_label(pred_GB[0]),
                                                                                                     output_label(pred_RF[0])))
    
```

```python

news = str(input("Enter the news text for testing: "))


manual_testing(news)
```
